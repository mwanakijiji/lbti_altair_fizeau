```python
# This reads in a single ADI frame with no fake companion, and writes out the
# frame one time for each radial value we want to sample. This allows the regular 
# pipeline to check the noise levels 

# Created 2019 July 9 by E.S.
```

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
```

```python
stem = "/Volumes/seagate_external_drive/lbti_data_reduction/180507_fizeau_altair/" + \
    "pipeline_testing_adi_no_fake_planet_optimization/"
```

```python
radii_to_test = 0.1*np.arange(1,19)
```

```python
radii_to_test
```

```python
frame, hdr = fits.getdata(stem + "adi_frame_0.0_0.0_0.0.fits", 0, header=True)
```

```python
plt.imshow(frame, origin="lower")
plt.show()
```

```python
for t in range(0,len(radii_to_test)):
    
    # change the header info
    hdr["RADASEC"] = radii_to_test[t]
    
    # and the file names
    fits.writeto(filename = stem + "multiples/adi_frame_0.0_"+\
                 str(format(radii_to_test[t], '.1f'))+"_0.0.fits",
                     data = frame,
                     header = hdr,
                     overwrite = True)
    print(t)
```
