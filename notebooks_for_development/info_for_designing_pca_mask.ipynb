```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda_over_D_pix = 9.46 # for 4 um
```

```python
pa_tot = 40 # total parallactic angle change of the dataset
fwhm_pix = 1.028*lambda_over_D_pix # FWHM of Airy PSF in LMIR pixels for 4 um
```

```python
fwhm_pix
```

```python
np.pi*(np.power(0.5*fwhm_pix,2))
```

```python
pix_radial_range = np.arange(1,101)
```

```python
pix_circumf_range = 2*np.pi*pix_radial_range
```

```python
fwhm_circumf_range = np.divide(pix_circumf_range,fwhm_pix)
```

```python
plt.plot(pix_radial_range,fwhm_circumf_range)
plt.xlabel("Radius (pix)")
plt.ylabel("Number of FWHM in circumference")
plt.show()
```

```python
num_degrees_pa_for_1_fwhm_passage = np.divide(360.,fwhm_circumf_range)
```

```python
# Airy ring positions in lambda/D

first_dark = 1.220
second_dark = 2.233
third_dark = 3.238
fourth_dark = 4.241
fifth_dark = 5.243

first_bright = 1.635
second_bright = 2.679
third_bright = 3.69
fourth_bright = 4.71
fifth_bright = 5.71
```

```python
plt.plot(pix_radial_range,num_degrees_pa_for_1_fwhm_passage)
plt.xlim([0,100])
plt.ylim([0,50])
plt.axhline(y=40, linestyle="--")
plt.axvline(x=first_dark*lambda_over_D_pix, linestyle="--")
plt.axvline(x=second_dark*lambda_over_D_pix, linestyle="--")
plt.axvline(x=third_dark*lambda_over_D_pix, linestyle="--")
plt.axvline(x=fourth_dark*lambda_over_D_pix, linestyle="--")
plt.axvline(x=fifth_dark*lambda_over_D_pix, linestyle="--")
plt.xlabel("Radius (pix)")
plt.ylabel("Number of degrees PA for 1 FWHM passage")
#plt.show()
plt.savefig("deg_pa_requirement.pdf")
```
