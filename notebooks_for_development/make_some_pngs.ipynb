```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import os
import gc
from astropy.io import fits
```

```python
stem = ('/home/../../media/unasemaje/Elements/lbti_data_reduction/'+\
        '180507_fizeau_altair/05_centered/')
```

```python
startFrame = 6547
endFrame = 7926
for framenum in range(startFrame,endFrame+1): #endFrame+1): # 83, 67108
            
    # raw data (except that readout glitch correction has been done)
    name_string = 'lm_180507_'+str("{:0>6d}".format(framenum))+'_nPCA164'
    img_string = stem + name_string + '.fits'
    
    # if FITS file exists in the first place
    N = 1 # do every Nth frame
    if ((np.mod(framenum,N) == 0) & os.path.isfile(img_string)): 
            
        # read in image
        sciImg, header = fits.getdata(img_string,0,header=True)
    
        # write out pngs
        image = plt.imshow(sciImg)
        plt.savefig(stem + 'png_thumbnails/' + name_string + '.png', overwrite = True)
        plt.close()
        del(image)
        #gc.collect()
    
        print(framenum)
```
