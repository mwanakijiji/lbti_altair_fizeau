```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#df = pd.read_csv("data/test_csv_file_all_iterations_morning_of_20191121.csv")
#df = pd.read_csv("data/test_csv_file_all_iterations_afternoon_of_20191121_number01.csv")
#df = pd.read_csv("data/test_csv_file_all_iterations_morning_of_191122_number_01.csv")
#df1 = pd.read_csv("data/test_csv_file_all_iterations_20191121_number02b.csv")
#df = pd.read_csv("data/test_csv_file_all_iterations_20191121_number03.csv")
df = pd.read_csv("data/test_csv_file_all_iterations_afternoon_of_20191124_number01.csv")
```

```python
df
```

```python
test_rad1_large = df.where(np.logical_and(df["rad_asec"] == 0.20,df["ampl_linear_norm_0"] == 0.001))
test_rad2_large = df.where(np.logical_and(df["rad_asec"] == 0.25,df["ampl_linear_norm_0"] == 0.001))
test_rad3_large = df.where(np.logical_and(df["rad_asec"] == 0.30,df["ampl_linear_norm_0"] == 0.001))
test_rad4_large = df.where(np.logical_and(df["rad_asec"] == 0.35,df["ampl_linear_norm_0"] == 0.001))

test_rad1_small = df.where(np.logical_and(df["rad_asec"] == 0.20,df["ampl_linear_norm_0"] == 0.0001))
test_rad2_small = df.where(np.logical_and(df["rad_asec"] == 0.25,df["ampl_linear_norm_0"] == 0.0001))
test_rad3_small = df.where(np.logical_and(df["rad_asec"] == 0.30,df["ampl_linear_norm_0"] == 0.0001))
test_rad4_small = df.where(np.logical_and(df["rad_asec"] == 0.35,df["ampl_linear_norm_0"] == 0.0001))
```

# Amplitude convergence

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6), sharey=True)

ax1.axhline(y=0, linestyle=":", color="k")
ax1.plot(test_rad1_small["inject_iteration"].dropna(),test_rad1_small["ampl_linear_norm"].dropna(), marker='o', label="0.20 asec")
ax1.plot(test_rad2_small["inject_iteration"].dropna(),test_rad2_small["ampl_linear_norm"].dropna(), marker='o', label="0.25 asec")
ax1.plot(test_rad3_small["inject_iteration"].dropna(),np.add(0.0005,test_rad3_small["ampl_linear_norm"].dropna()), marker='o', label="0.30 asec (+ y-offset)")
ax1.plot(test_rad4_small["inject_iteration"].dropna(),test_rad4_small["ampl_linear_norm"].dropna(), marker='o', label="0.35 asec")
ax1.set_ylim([-0.001,0.005])
ax1.set_xlim([-1,12])
ax1.set_xlabel("Step")
ax1.set_ylabel("Amplitude")
ax1.set_title("Amplitude convergence, 1e-4 ampl start")
ax1.legend()

ax2.axhline(y=0, linestyle=":", color="k")
ax2.plot(test_rad1_large["inject_iteration"].dropna(),test_rad1_large["ampl_linear_norm"].dropna(), marker='o', label="0.20 asec")
ax2.plot(test_rad2_large["inject_iteration"].dropna(),test_rad2_large["ampl_linear_norm"].dropna(), marker='o', label="0.25 asec")
ax2.plot(test_rad3_large["inject_iteration"].dropna(),np.add(0.0005,test_rad3_large["ampl_linear_norm"].dropna()), marker='o', label="0.30 asec (+ y-offset)")
ax2.plot(test_rad4_large["inject_iteration"].dropna(),test_rad4_large["ampl_linear_norm"].dropna(), marker='o', label="0.35 asec")
ax2.set_ylim([-0.001,0.005])
ax2.set_xlim([-1,12])
ax2.set_xlabel("Step")
#ax2.set_ylabel("Planet amplitude (relative to host star)")
ax2.set_title("Amplitude convergence, 1e-3 ampl start")
ax2.legend()

f.show()
#plt.savefig("junk_convergence_functional_191124.pdf")
```

# S/N convergence

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6), sharey=True)

ax1.axhline(y=5, linestyle=":", color="k")
ax1.plot(test_rad1_small["inject_iteration"].dropna(),test_rad1_small["s2n"].dropna(), marker='o', 
         label="0.20 asec (final S/N=5.008)")
ax1.plot(test_rad2_small["inject_iteration"].dropna(),test_rad2_small["s2n"].dropna(), marker='o', 
         label="0.25 asec (final S/N=5.141)")
ax1.plot(test_rad3_small["inject_iteration"].dropna(),test_rad3_small["s2n"].dropna(), marker='o', 
         label="0.30 asec (final S/N=4.466)")
ax1.plot(test_rad4_small["inject_iteration"].dropna(),test_rad4_small["s2n"].dropna(), marker='o', 
         label="0.35 asec (final S/N=4.181)")
#ax1.set_ylim([-0.001,0.005])
ax1.set_xlabel("Step")
ax1.set_ylabel("S/N")
ax1.set_title("S/N convergence, 1e-4 ampl start")
ax1.legend()

ax2.axhline(y=5, linestyle=":", color="k")
ax2.plot(test_rad1_large["inject_iteration"].dropna(),test_rad1_large["s2n"].dropna(), marker='o', 
         label="0.20 asec (final S/N=5.008)")
ax2.plot(test_rad2_large["inject_iteration"].dropna(),test_rad2_large["s2n"].dropna(), marker='o', 
         label="0.25 asec (final S/N=4.979)")
ax2.plot(test_rad3_large["inject_iteration"].dropna(),test_rad3_large["s2n"].dropna(), marker='o', 
         label="0.30 asec (final S/N=4.466)")
ax2.plot(test_rad4_large["inject_iteration"].dropna(),test_rad4_large["s2n"].dropna(), marker='o', 
         label="0.35 asec (final S/N=4.181)")
#ax2.set_ylim([-0.001,0.005])
ax2.set_xlabel("Step")
#ax2.set_ylabel("Planet amplitude (relative to host star)")
ax2.set_title("S/N convergence, 1e-3 ampl start")
ax2.legend()

f.show()
#plt.savefig("junk_s2n_convergence_191124.pdf")
```
